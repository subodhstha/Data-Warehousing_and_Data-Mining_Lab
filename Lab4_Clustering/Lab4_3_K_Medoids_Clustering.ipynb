```python
# Import necessary libraries
import pandas as pd
import numpy as np
from pyclustering.cluster.kmedoids import kmedoids
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Actual labels (for evaluation, not used in clustering)

# Initialize the K-Medoids model with 3 clusters
initial_medoids = [0, 50, 100]  # Initial medoids (arbitrary initialization)
kmedoids_model = kmedoids(X, initial_medoids)

# Fit the model
kmedoids_model.process()

# Get the cluster labels
y_pred = kmedoids_model.get_clusters()

# Get the final medoids (centers of the clusters)
final_medoids = kmedoids_model.get_medoids()

# Print the final medoids
print("Final Medoids:\n", final_medoids)

# Visualizing the clusters (using the first two features for 2D visualization)
plt.figure(figsize=(8, 6))

# Plot the data points, colored by predicted cluster labels
for cluster in y_pred:
    plt.scatter(X[cluster, 0], X[cluster, 1], label=f"Cluster {y_pred.index(cluster)}")

# Plot the final medoids (centers)
for medoid in final_medoids:
    plt.scatter(X[medoid, 0], X[medoid, 1], c='red', marker='X', s=200, label="Medoid")

# Add labels and title
plt.title("K-Medoids Clustering on Iris Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
