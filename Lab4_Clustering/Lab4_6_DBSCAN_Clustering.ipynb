```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Actual labels (for evaluation, not used in clustering)

# Create a DBSCAN model
dbscan = DBSCAN(eps=0.5, min_samples=5)

# Fit the model and predict the cluster labels
y_pred = dbscan.fit_predict(X)

# Print the predicted cluster labels
print("Predicted Cluster Labels:\n", y_pred)

# Visualizing the clusters (using the first two features for 2D visualization)
plt.figure(figsize=(8, 6))

# Plot each cluster
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis')

# Add labels and title
plt.title("DBSCAN Clustering on Iris Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

# Show the plot
plt.show()
```
