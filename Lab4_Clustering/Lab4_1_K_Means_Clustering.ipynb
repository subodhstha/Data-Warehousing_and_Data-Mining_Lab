```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Actual labels (for evaluation, not used in clustering)

# Create a KMeans clustering model with 3 clusters (since we know there are 3 types of iris flowers)
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model to the data
kmeans.fit(X)

# Predict the cluster labels
y_pred = kmeans.predict(X)

# Get the cluster centers
centers = kmeans.cluster_centers_

# Print the cluster centers
print("Cluster Centers:\n", centers)

# Visualizing the clusters (using the first two features for 2D visualization)
plt.figure(figsize=(8, 6))

# Scatter plot of data points, colored by predicted cluster labels
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', label='Data Points')

# Plot the cluster centers
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Cluster Centers')

# Add labels and title
plt.title("K-Means Clustering on Iris Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
