```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset (focusing on "Marks")
data = {
    "Marks": [45, 60, 75, 80, 65, 50, 85, 90, 40, 95],
    "Age": [20, 21, 22, 23, 20, 21, 22, 23, 19, 24],
    "Study Hours": [2, 4, 6, 8, 5, 3, 7, 9, 1, 10]
}

df = pd.DataFrame(data)

# Scatter Plot: Marks vs Study Hours
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Study Hours", y="Marks", data=df)
plt.title("Scatter Plot: Marks vs Study Hours")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()

# Histogram: Distribution of Marks
plt.figure(figsize=(8, 5))
sns.histplot(df["Marks"], bins=5, kde=True, color='blue')
plt.title("Histogram: Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# Box Plot: Marks vs Age
plt.figure(figsize=(8, 5))
sns.boxplot(x="Age", y="Marks", data=df)
plt.title("Box Plot: Marks vs Age")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()
```
